```python
import os
import numpy as np
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.models import load_model
from PIL import Image, ImageDraw
import matplotlib.pyplot as plt
```

```python
current_directory = os.path.dirname(os.path.realpath(r"C:\Users\Admin\Desktop\CapstoneProject\capstone\mainFolder\predictinIMG.ipynb"))

# Define the relative path to the trained model
model_path = os.path.join(current_directory, 'object_detection_model.h5')

# Load the trained model
model = load_model(model_path)


```

```python
new_image_path = r'C:\Users\Admin\Desktop\CapstoneProject\capstone\people.jpeg'  # Replace with the actual path to your new dark image
```

```python
img = load_img(new_image_path, target_size=(224,224))
img_array = img_to_array(img)
img_array = np.expand_dims(img_array, axis=0) / 255.0   
```

```python
class_names = ['Bicycle', 'Boat', 'Bottle', 'Bus', 'Car', 'Cat', 'Chair', 'Cup', 'Dog', 'Moterbike', 'People', 'Table']
```

```python
predictions = model.predict(img_array)
# Get the predicted class name
predicted_class_index = np.argmax(predictions)
predicted_class_name = class_names[predicted_class_index]

plt.imshow(img)
plt.title(f"Predicted Class: {predicted_class_name}")
plt.axis('off')
plt.show()
```
